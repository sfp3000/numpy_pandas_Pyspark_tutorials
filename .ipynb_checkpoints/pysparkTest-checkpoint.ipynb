##### Selecting columns

```python
df.select("id", "date", 'close', 'macd1').show(10)
```

```python
print(df.select(min('macd1')).collect()[0])
```

```python
#Filter for closing price > 20
df.filter(col("Close") > 20).show(20)
```

```python
df.where("Close > 20").show()
```

```python
# Adding new columns
```

```python
df = df.withColumn("Daily_Return", (col("Close") - col("Open")) /col("Open"))
```

```python
df.show(5)
```

```python
#grouping and Aggregation
#Monthly average closing price
df.groupBy(month("Date").alias("Month")) \
    .agg(avg("Close").alias("Avg_Close")) \
    .orderBy("Month").show(20)
```

```python
# handling missing data
df_clean = df.na.drop()
```

```python
#fill nulls with mean
from pyspark.sql.functions import mean
```

```python
mean_close = df.select(mean("Close")).collect()[0][0]
df_filled = df.na.fill(mean_close, subset=['Close'])
```

```python
min_close = df.select(min("Close")).collect()[0][0]
```

```python
min_close
```

```python
mean_close
```

```python
#Advanced Techniques 
#Window functions
```

```python
from pyspark.sql.window import Window
window_spec = Window.orderBy("Date").rowsBetween(-7, 0)
df = df.withColumn("7Day_MA", avg("Close").over(window_spec))
```

```python
df.show(50)
```

```python
df.select("id", "date", 'close', 'macd1', "7Day_MA").show(10)
```

```python
#UDFs (User-Defined Functions)
#Register a Python function as UDF
def categorize_vol(volume):
    return "High" if volume > 1e7 else "Low"

vol_udf = udf(categorize_vol, StringType())
df = df.withColumn("Volume_Category", vol_udf(col("Volume")))
```

```python
df
```

```python
df.show(5)
```

```python
df.select("Volume_Category").show(5)
```

```python
#Performance Optimization
```

```python
#Cache frequently used DataFrames
df.cache()
```

```python
#Repartition for better parallelism
df = df.repartition(4)
```

```python
df
```

```python
# Check execution plan
df.explain()
```

```python
#Saving results:
#to parquet (columnar format):
df.write.parquet("intc_processed.parquet")
```

```python
#exporting to csv
df.write.csv("intc_processed.csv", header=True)
```
