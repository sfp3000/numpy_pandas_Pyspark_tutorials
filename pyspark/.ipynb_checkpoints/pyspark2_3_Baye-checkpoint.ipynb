```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB

import findspark
findspark.init(r"C:\python311\Lib\site-packages\pyspark")
```

```python
def generate_data(n_samples=200, random_state=42):
    """
    Generate a synthetic binary classification dataset.
    Returns: X (features), y (labels)
    """
    X, y = make_classification(
        n_samples=n_samples,
        n_features=2,       # keep 2D for easy plotting
        n_classes=2,
        n_informative=2,
        n_redundant=0,
        random_state=random_state
    )
    return X, y

```

```python
def plot_data_and_model(X, y, model):
    """
    Plot the dataset and the decision boundary of the trained Bayesian model.
    """
    plt.figure(figsize=(8,6))
    plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')

    # Decision boundary grid
    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.2, cmap=plt.cm.coolwarm)
    plt.title("Binary Classification with Gaussian Naive Bayes")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

```

```python
def train_bayesian_model(X, y):
    """
    Train a Gaussian Naive Bayes classifier and call the plotting function.
    """
    model = GaussianNB()
    model.fit(X, y)
    plot_data_and_model(X, y, model)
    return model
```

```python

# Example workflow:
X, y = generate_data()
model = train_bayesian_model(X, y)
```
