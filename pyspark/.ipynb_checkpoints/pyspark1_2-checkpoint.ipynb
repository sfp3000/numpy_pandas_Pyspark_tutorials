```python
import numpy as np

def generate_linear_regression_data(n_samples=100, slope=2.5, intercept=5, noise_std=1.0):
    """
    Generate synthetic data for a linear regression lesson.
    
    Parameters:
    - n_samples (int): Number of data points
    - slope (float): True slope of the line
    - intercept (float): True intercept of the line
    - noise_std (float): Standard deviation of Gaussian noise
    
    Returns:
    - X (numpy array): Feature values (shape: [n_samples, 1])
    - y (numpy array): Target values (shape: [n_samples, ])
    """
    # Generate random feature values
    X = np.linspace(0, 10, n_samples).reshape(-1, 1)
    
    # Generate target values with linear relationship + noise
    y = slope * X.flatten() + intercept + np.random.normal(0, noise_std, n_samples)
    
    return X, y
```

```python
import numpy as np

def generate_linear_regression_data_ds(n_samples=100, n_features=1, noise=0.1, random_state=None):
    """
    Generate random data for linear regression.
    
    Parameters:
    n_samples (int): Number of samples
    n_features (int): Number of features
    noise (float): Standard deviation of Gaussian noise
    random_state (int): Random seed for reproducibility
    
    Returns:
    tuple: (X, y) where:
        - X: numpy array of shape (n_samples, n_features) with features
        - y: numpy array of shape (n_samples,) with target values
    """
    if random_state is not None:
        np.random.seed(random_state)
    
    # Generate random features
    X = np.random.randn(n_samples, n_features)
    
    # Generate random coefficients for the linear relationship
    coefficients = np.random.randn(n_features)
    
    # Generate intercept
    intercept = np.random.randn()
    
    # Calculate target values with linear relationship and noise
    y = intercept + X.dot(coefficients) + noise * np.random.randn(n_samples)
    
    return X, y
```

```python
import numpy as np

def generate_linear_regression_dataset_ds2(n_samples=100, n_features=1, noise=0.1, random_state=None):
    """
    Generate random data for linear regression.
    
    Parameters:
    n_samples (int): Number of samples
    n_features (int): Number of features
    noise (float): Standard deviation of Gaussian noise
    random_state (int): Random seed for reproducibility
    
    Returns:
    numpy array: Combined dataset of shape (n_samples, n_features + 1)
    """
    if random_state is not None:
        np.random.seed(random_state)
    
    # Generate random features
    X = np.random.randn(n_samples, n_features)
    
    # Generate random coefficients for the linear relationship
    coefficients = np.random.randn(n_features)
    
    # Generate intercept
    intercept = np.random.randn()
    
    # Calculate target values with linear relationship and noise
    y = intercept + X.dot(coefficients) + noise * np.random.randn(n_samples)
    
    # Combine into single numpy array
    dataset = np.column_stack((X, y))
    
    return dataset
```

```python
import numpy as np

def generate_linear_dataset_copilot(n_samples=100, slope=2.0, bias=5.0, noise_std=1.0):
    """
    Generate a synthetic linear regression dataset as a single NumPy array.
    
    Each row = [feature, target]
    
    Parameters:
    - n_samples (int): Number of data points
    - slope (float): Line slope (weight)
    - bias (float): Line intercept (bias term)
    - noise_std (float): Standard deviation of Gaussian noise
    
    Returns:
    - dataset (numpy array): Shape (n_samples, 2)
    """
    # Random feature values between 0 and 10
    X = np.random.uniform(0, 10, n_samples)
    
    # Linear relationship with noise
    y = slope * X + bias + np.random.normal(0, noise_std, n_samples)
    
    # Combine into one dataset: column 0 = X, column 1 = y
    dataset = np.column_stack((X, y))
    
    return dataset
```

# Linear Regression with PySpark

```python
import numpy as np
 
def generate_linear_dataset_copilot(n_samples=100, slope=2.0, bias=5.0, noise_std=1.0):
    """
    Generate a synthetic linear regression dataset as a single NumPy array.
    
    Each row = [feature, target]
    
    Parameters:
    - n_samples (int): Number of data points
    - slope (float): Line slope (weight)
    - bias (float): Line intercept (bias term)
    - noise_std (float): Standard deviation of Gaussian noise
    
    Returns:
    - dataset (numpy array): Shape (n_samples, 2)
    """
    # Random feature values between 0 and 10
    X = np.random.uniform(0, 10, n_samples)
    
    # Linear relationship with noise
    y = slope * X + bias + np.random.normal(0, noise_std, n_samples)
    
    # Combine into one dataset: column 0 = X, column 1 = y
    dataset = np.column_stack((X, y))
    
    return dataset
```

```python
#SOLUTION 1
from pyspark.sql import SparkSession
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
import matplotlib.pyplot as plt
import numpy as np

def pyspark_linear_regression(dataset):
    """
    Perform linear regression using PySpark and visualize results.
    
    Parameters:
    dataset: numpy array from generate_linear_dataset_copilot
    """
    # Create Spark session
    spark = SparkSession.builder \
        .appName("LinearRegression") \
        .getOrCreate()
    
    # Convert numpy array to Spark DataFrame
    # dataset has columns: [feature, target]
    spark_df = spark.createDataFrame(dataset.tolist(), ["feature", "target"])
    
    # Prepare features for MLlib (requires vector column)
    assembler = VectorAssembler(inputCols=["feature"], outputCol="features")
    df = assembler.transform(spark_df)
    
    # Create and train linear regression model
    lr = LinearRegression(featuresCol="features", labelCol="target")
    model = lr.fit(df)
    
    # Make predictions
    predictions = model.transform(df)
    
    # Get model coefficients
    coefficients = model.coefficients
    intercept = model.intercept
    
    print(f"PySpark Linear Regression Results:")
    print(f"Slope (coefficient): {coefficients[0]:.4f}")
    print(f"Intercept: {intercept:.4f}")
    print(f"R: {model.summary.r2:.4f}")
    
    # Convert back to pandas for visualization
    pandas_df = predictions.toPandas()
    
    # Visualize results
    plt.figure(figsize=(10, 6))
    
    # Plot original data
    plt.scatter(pandas_df['feature'], pandas_df['target'], alpha=0.7, 
                label='Original Data', color='blue')
    
    # Plot regression line
    x_vals = np.linspace(pandas_df['feature'].min(), pandas_df['feature'].max(), 100)
    y_vals = coefficients[0] * x_vals + intercept
    plt.plot(x_vals, y_vals, color='red', linewidth=2, 
             label=f'Regression Line: y = {coefficients[0]:.2f}x + {intercept:.2f}')
    
    plt.xlabel('Feature')
    plt.ylabel('Target')
    plt.title('PySpark Linear Regression')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # Stop Spark session
    spark.stop()
    
    return model

# Complete example usage:
if __name__ == "__main__":
    # Generate data
    dataset = generate_linear_dataset_copilot(n_samples=100, slope=2.0, bias=5.0, noise_std=1.0)
    
    # Run PySpark linear regression and visualize
    model = pyspark_linear_regression(dataset)
```

```python
#SOLUTION 2
import numpy as np
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession, Row
from pyspark.ml.linalg import Vectors
from pyspark.ml.regression import LinearRegression

# -------------------------------
# Function 1: Create data with NumPy
# -------------------------------
def generate_linear_dataset(n_samples=100, slope=2.0, bias=5.0, noise_std=1.0):
    """
    Generate synthetic linear regression dataset using NumPy.
    Each row = [X, y]
    """
    X = np.random.uniform(0, 10, n_samples)
    y = slope * X + bias + np.random.normal(0, noise_std, n_samples)
    return np.column_stack((X, y))

# -------------------------------
# Function 2: Visualize dataset
# -------------------------------
def visualize_dataset(dataset, model=None):
    """
    Visualize dataset with optional regression line.
    
    Parameters:
    - dataset: NumPy array with rows [X, y]
    - model: PySpark LinearRegressionModel (optional)
    """
    X = dataset[:, 0]
    y = dataset[:, 1]
    
    plt.scatter(X, y, color="blue", alpha=0.6, label="Data points")
    
    # If model is provided, plot regression line
    if model is not None:
        y_pred = model.coefficients[0] * X + model.intercept
        plt.plot(X, y_pred, color="red", linewidth=2, label=f'Regression Line: y = {model.coefficients[0]:.2f}x + {model.intercept:.2f}')
        # plt.plot(X, y_pred, color='red', linewidth=2, 
        #      label=f'Regression Line: y = {coefficients[0]:.2f}x + {intercept:.2f}')
    
    plt.xlabel("X (feature)")
    plt.ylabel("y (target)")
    plt.title("Linear Regression Visualization")
    plt.legend()
    plt.show()

# -------------------------------
# Function 3: Fit PySpark model and visualize
# -------------------------------
def fit_model_pyspark(data_function, **kwargs):
    """
    Fit a PySpark LinearRegression model on dataset generated by data_function.
    Automatically calls visualization.
    
    Parameters:
    - data_function: function that generates dataset (e.g., generate_linear_dataset)
    - kwargs: arguments passed to data_function
    
    Returns:
    - model: Trained PySpark LinearRegressionModel
    """
    # Step 1: Generate dataset by calling the function
    dataset = data_function(**kwargs)
    
    # Step 2: Initialize Spark
    spark = SparkSession.builder.getOrCreate()
    
    # Step 3: Convert NumPy dataset into Spark DataFrame
    data = [Row(features=Vectors.dense([float(x)]), label=float(y)) for x, y in dataset]
    df = spark.createDataFrame(data)
    
    # Step 4: Fit PySpark Linear Regression
    lr = LinearRegression(featuresCol="features", labelCol="label")
    model = lr.fit(df)
    
    # Step 5: Visualize results (calls Function 2)
    visualize_dataset(dataset, model)
    print(model)
    
    return model
```

```python
#Chat GPT
import numpy as np

def generate_dataset(n_samples=100, dataset_type="linear_regression", 
                     slope=1.0, intercept=0.0, noise_std=1.0, 
                     n_clusters=3, cluster_std=1.0):
    """
    Generates synthetic datasets for machine learning.

    Parameters:
        n_samples (int): number of samples
        dataset_type (str): "linear_regression" or "clustering"
        slope (float): slope for linear regression data
        intercept (float): intercept for linear regression data
        noise_std (float): noise standard deviation
        n_clusters (int): number of clusters for clustering data
        cluster_std (float): cluster spread for clustering data
        
    Returns:
        np.ndarray: dataset
    """

    if dataset_type == "linear_regression":
        # X: evenly spaced numbers
        X = np.linspace(0, 10, n_samples)
        # y = slope * x + intercept + noise
        noise = np.random.normal(0, noise_std, n_samples)
        y = slope * X + intercept + noise
        return np.column_stack((X, y))
    
    elif dataset_type == "clustering":
        # Create cluster centers randomly
        centers = np.random.uniform(-10, 10, size=(n_clusters, 2))
        data = []

        # Assign points around each center
        for center in centers:
            cluster_points = center + np.random.randn(n_samples // n_clusters, 2) * cluster_std
            data.append(cluster_points)

        return np.vstack(data)

    else:
        raise ValueError("Unsupported dataset_type. Use 'linear_regression' or 'clustering'.")
```
