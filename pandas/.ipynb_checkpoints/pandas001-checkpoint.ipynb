# 10 04 Pandas Data Cleaning & Transformation

```python
import pandas as pd
```

```python
df = pd.read_csv('intc.csv', index_col='id')
```

```python
df.head(3)
```

```python
# Handling Missing Data with Pandas
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
#drop specific columns only
df.dropna(subset=['close', 'open'])
```

```python
#Fill missing data
```

```python
df.fillna(0)
```

```python
df.fillna({'close':0, 'macd1': 'missing'})
```

```python
df.fillna(method='ffill').head(3)
```

```python
df.fillna(method='bfill').head(3)
```

```python
df.interpolate()
```

```python
# Removing duplicates
df.drop_duplicates()   #Drops exact duplicates
```

```python
df.drop_duplicates(subset=['date']).head(3)
```

```python
df.drop_duplicates(keep='last').head(3)
```

```python
df.dtypes
```

```python
df['ema10'] = df['ema10'].astype('int8')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df['mystring'] = df['mystring'].str.capitalize()
```
